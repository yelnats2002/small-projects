# Exploratory Data Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a=pd.read_csv('Final dataset Attrition (1).csv')
```

```python
a.head()
```

```python
a.dtypes
```

```python
a['Date_of_Hire'] = pd.to_datetime(a['Date_of_Hire'], format='%d/%m/%Y')
```

```python
a['Date_of_Hire'].dtype
```

```python
a.isnull().sum()
```

```python
a.drop(['Date_of_termination','Unnamed: 32'],axis=1)
```

```python
a.shape
```

```python
a.columns
```

```python
plt.hist(a['Age'],bins=10)
plt.title('HISTOGRAM ON THE BASIS ON AGE')
plt.show()
```

```python
Attrition=a['Attrition'].value_counts()
fig,ax = plt.subplots()
ax.pie(Attrition.values,labels=Attrition.index,autopct='%1.1f%%')
ax.axis('equal')
ax.set_title("Attrition percentage")
plt.show()
```

```python
BusinessTravel=a['BusinessTravel'].value_counts()
plt.bar(BusinessTravel.index,BusinessTravel.values)
plt.title('Business Travel Frequency')
plt.ylabel('count of travel')
```

```python
c=a['Department'].value_counts().index
d=a['Department'].value_counts().values
plt.bar(c,d,width=0.2)
plt.title('Number of employee in different department')
plt.xlabel('')
plt.ylabel('')
```

```python
b=a['Status_of_leaving'].value_counts()
fig, ax = plt.subplots()
ax.pie(b.values, labels=b.index, autopct='%1.1f%%')
ax.axis('equal')
plt.title('Reason for employee leaving the company')
plt.show()
```

```python
plt.figure(figsize=(20,10))
c=a['JobRole'].value_counts().values
d=a['JobRole'].value_counts().index
plt.ylabel('count of employee')
plt.title('Number of employee in each job role')
plt.bar(d,c)
```

```python
sns.scatterplot(x=a['TotalWorkingYears'],y=a['MonthlyIncome'])
plt.title('Finding outlier on basis of salary')
```

```python
import plotly.express as px
fig = px.density_heatmap(a, x='Mode_of_work', y='DistanceFromHome')
fig.update_layout(title='Distance from Home by Work Mode')
fig.update_xaxes(title='Work Mode')
fig.update_yaxes(title='Distance from Home')
fig.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=4, figsize=(12, 4))
a.groupby('Job_mode')['Job_mode'].count().plot(kind='bar', ax=ax[0])
ax[0].set_xlabel('Job Mode')
ax[0].set_ylabel('Count')
ax[0].set_title('Job Mode')

a.groupby('Higher_Education')['Higher_Education'].count().plot(kind='bar', ax=ax[1])
ax[1].set_xlabel('Higher Education')
ax[1].set_ylabel('Count')
ax[1].set_title('Higher Education')

a.groupby('Mode_of_work')['Mode_of_work'].count().plot(kind='bar', ax=ax[2])
ax[2].set_xlabel('Mode of Work')
ax[2].set_ylabel('Count')
ax[2].set_title('Mode of Work')

a.groupby('Source_of_Hire')['Source_of_Hire'].count().plot(kind='bar', ax=ax[3])
ax[3].set_xlabel('Source_of_Hire')
ax[3].set_ylabel('Count')
ax[3].set_title('Source_of_Hire')

plt.subplots_adjust(wspace=0.5)

plt.show()
```
